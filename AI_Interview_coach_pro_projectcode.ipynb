```python
# ================================
# AI INTERVIEW COACH PRO
# Complete Google Colab Project
# ================================

!pip -q install PyPDF2

from PyPDF2 import PdfReader
from google.colab import files
import random
import pandas as pd

# ================================
# ATS Keywords
# ================================

ATS_KEYWORDS = [
    "python","java","c","c++","sql",
    "html","css","javascript","react",
    "node","git","github","tensorflow",
    "opencv","machine learning",
    "deep learning","communication",
    "leadership","teamwork",
    "problem solving","project",
    "internship"
]

# ================================
# PDF Text Extraction
# ================================

def extract_text(pdf_file):
    text = ""

    reader = PdfReader(pdf_file)

    for page in reader.pages:
        page_text = page.extract_text()

        if page_text:
            text += page_text

    return text.lower()

# ================================
# ATS Score
# ================================

def ats_score(text):

    found = []

    for keyword in ATS_KEYWORDS:
        if keyword in text:
            found.append(keyword)

    score = int((len(found)/len(ATS_KEYWORDS))*100)

    return score, found

# ================================
# Missing Skills
# ================================

def missing_keywords(text):

    missing = []

    for keyword in ATS_KEYWORDS:

        if keyword not in text:
            missing.append(keyword)

    return missing

# ================================
# Skill Detection
# ================================

def detect_skills(text):

    skills = []

    skill_database = [
        "python",
        "java",
        "c",
        "c++",
        "sql",
        "html",
        "css",
        "javascript",
        "react",
        "node",
        "tensorflow",
        "opencv",
        "machine learning",
        "deep learning"
    ]

    for skill in skill_database:

        if skill in text:
            skills.append(skill)

    return skills

# ================================
# HR Questions
# ================================

HR_QUESTIONS = [

    "Tell me about yourself.",

    "Why should we hire you?",

    "What are your strengths?",

    "What are your weaknesses?",

    "Describe a challenge you faced.",

    "How do you handle pressure?",

    "Where do you see yourself in 5 years?",

    "Why do you want this job?",

    "Tell me about your leadership experience.",

    "What motivates you?"
]

# ================================
# Technical Questions
# ================================

TECH_DB = {

    "python":[
        "What are decorators in Python?",
        "Difference between list and tuple?",
        "What are generators?"
    ],

    "sql":[
        "Explain SQL joins.",
        "Difference between DELETE and TRUNCATE?",
        "What is normalization?"
    ],

    "java":[
        "What is JVM?",
        "Explain inheritance.",
        "Difference between JDK and JRE?"
    ],

    "machine learning":[
        "What is overfitting?",
        "What is supervised learning?",
        "Explain cross validation."
    ],

    "opencv":[
        "What is image thresholding?",
        "Explain edge detection.",
        "Difference between RGB and BGR?"
    ]
}

# ================================
# Project Questions
# ================================

PROJECT_QUESTIONS = [

    "Explain your project architecture.",

    "What challenges did you face?",

    "How did you test your project?",

    "Why did you choose this technology?",

    "What improvements can be made?",

    "What was your contribution?",

    "How would you scale this project?"
]

# ================================
# Interview Readiness Score
# ================================

def readiness_score(ats, skills):

    score = ats * 0.7 + len(skills) * 3

    return min(100, int(score))

# ================================
# Answer Evaluation
# ================================

def evaluate_answer(answer):

    words = len(answer.split())

    if words < 20:
        return "Weak Answer"

    elif words < 50:
        return "Average Answer"

    else:
        return "Strong Answer"

# ================================
# MAIN PROGRAM
# ================================

print("="*60)
print("AI INTERVIEW COACH PRO")
print("="*60)

print("\nUpload Resume PDF")

uploaded = files.upload()

filename = list(uploaded.keys())[0]

text = extract_text(filename)

ats, found_keywords = ats_score(text)

skills = detect_skills(text)

missing = missing_keywords(text)

readiness = readiness_score(ats, skills)

# ================================
# RESULTS
# ================================

print("\n")
print("="*60)
print("RESUME ANALYSIS")
print("="*60)

print(f"ATS SCORE : {ats}/100")
print(f"INTERVIEW READINESS : {readiness}/100")

print("\nDetected Skills:")
print(skills)

print("\nKeywords Found:")
print(found_keywords)

print("\nTop Missing Keywords:")
print(missing[:10])

# ================================
# HR QUESTIONS
# ================================

print("\n")
print("="*60)
print("HR INTERVIEW QUESTIONS")
print("="*60)

for q in random.sample(HR_QUESTIONS,5):
    print("•",q)

# ================================
# TECHNICAL QUESTIONS
# ================================

print("\n")
print("="*60)
print("TECHNICAL QUESTIONS")
print("="*60)

tech_questions = []

for skill in skills:

    if skill in TECH_DB:
        tech_questions.extend(TECH_DB[skill])

if len(tech_questions)==0:

    tech_questions.append(
        "Explain your final year project."
    )

for q in tech_questions:
    print("•",q)

# ================================
# PROJECT QUESTIONS
# ================================

print("\n")
print("="*60)
print("PROJECT QUESTIONS")
print("="*60)

for q in PROJECT_QUESTIONS:
    print("•",q)

# ================================
# MOCK INTERVIEW
# ================================

print("\n")
print("="*60)
print("MOCK INTERVIEW")
print("="*60)

question = random.choice(HR_QUESTIONS)

print("\nQuestion:")
print(question)

answer = input("\nYour Answer:\n")

result = evaluate_answer(answer)

print("\nEvaluation :",result)

# ================================
# REPORT
# ================================

report = pd.DataFrame({

    "Metric":[
        "ATS Score",
        "Interview Readiness",
        "Skills Found"
    ],

    "Value":[
        ats,
        readiness,
        len(skills)
    ]
})

print("\n")
print("="*60)
print("FINAL REPORT")
print("="*60)

print(report)

report.to_csv(
    "Interview_Report.csv",
    index=False
)

print("\nInterview_Report.csv saved successfully.")
```
